# 4. Create a Loyalty Flag for existing customers

```python
# Create new column 'max_order'

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['number_of_orders'].transform(np.max)
```

```python
# Check the new column

ords_prods_merge.head(20)
```

```python
# Loyalty Flag 'Loyal Customer'

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Loyalty Flag 'Regular Customer'

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Loyalty Flag 'New Customer'

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check Loyalty Flags

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check the output -> Column 'Loyalty Flag'

ords_prods_merge[['user_id', 'loyalty_flag', 'number_of_orders']].head(20)
```

# 5. Difference in spending habits for the three types of customers

```python
# Group loyalty flag customers by the min, max, and mean prices

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Loyal customers are spending the least, while new customers are spending the most.
The regular customers are in between.

There is a clear outlier for the maximum price value.
For this reason, we have to have them marked as missing value.

```python
# Mark outliers as missing data

ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
# Check the output

ords_prods_merge['prices'].max()
```

```python
# Group loyalty flag customers by the min, max, and mean prices

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

# 6. Create a Spending Flag for users based on the avg. price basket

```python
# Create average_spend column

ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking the new column

ords_prods_merge.head(20)
```

```python
# Flag for 'Low Spender'

ords_prods_merge.loc[ords_prods_merge['average_spend']<10,'spender_flag']='low_spender'
```

```python
# Flag for 'High Spender'

ords_prods_merge.loc[ords_prods_merge['average_spend']>= 10, 'spender_flag'] = 'high_spender'
```

```python
# Check Spender Flags

ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# Check the output

ords_prods_merge[['user_id', 'average_spend', 'spender_flag', 'prices']].head(60)
```

# 7. Create a Flag for Frequent Users

```python
# Create a column for Frequent Users
# Group data by 'user_id'
# Generate median product of 'days_since_prior_order'
# Create 'customer_frequency' to aggregate the results


ords_prods_merge['customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check the new column

ords_prods_merge.head(20)
```

```python
# Flag for 'Non-Frequent Customer'

ords_prods_merge.loc[ords_prods_merge['customer_frequency'] > 20, 'frequency_flag'] = 'non_frequent_customer'
```

```python
# Flag for 'Regular Customer'

ords_prods_merge.loc[(ords_prods_merge['customer_frequency']>10) & (ords_prods_merge['customer_frequency']<=20), 'frequency_flag'] = 'regular_customer'
```

```python
# Flag for 'Frequent Customer'

ords_prods_merge.loc[ords_prods_merge['customer_frequency'] <= 10, 'frequency_flag'] = 'frequent_customer'
```

```python
# Check Frequency Flags

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
#Check new output

ords_prods_merge[['user_id', 'customer_frequency', 'frequency_flag']].head(60)
```
